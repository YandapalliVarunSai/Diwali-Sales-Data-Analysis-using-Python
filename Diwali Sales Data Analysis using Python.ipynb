```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt # visualizing data
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv(r'Downloads/Python_Diwali_Sales_Analysis-main/Python_Diwali_Sales_Analysis-main/Diwali Sales Data.csv',encoding="unicode_escape")
```

```python
df.shape
```

```python
df.head()
```

```python
df.info()
```

```python
#drop unrelated/blank columns
df.drop(['Status','unnamed1'],axis=1,inplace=True)
```

```python
#check for null values 
pd.isnull(df).sum()
```

```python
#drop null values
df.dropna(inplace=True)
```

```python
#change data type
df['Amount']=df['Amount'].astype('int')
```

```python
df['Amount'].dtypes
```

```python
df.columns
```

```python
#rename columns
df.rename(columns={'Marital_Status':'Shaadi'})
```

```python
#describe() method returns the description of the data in the dataframe(mean,median,std..) 
df.describe()
```

<h1>Exploratory Data Analysis</h1>

<h2>Gender</h2>

```python
# plotting a barchart for Gender and it's count
ax=sns.countplot(x='Gender',data=df,hue="Gender") 

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#plotting a bar graph between gender vs total amount 
sales_gen=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by="Amount",ascending=False) 
sns.barplot(x="Gender",y="Amount",data=sales_gen)
```

<p>From the above graph we can see that most of the buyers are female and even the purchasing power of female is higher than the men.</p>

<h1>Age</h1>

```python
ax=sns.countplot(data=df,x="Age Group",hue="Gender") 
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Total Amount vs Age Group
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Age Group',y= 'Amount' ,data = sales_age)
```

<p>From the above graphs most of the buyers are between a age group 26-35 females</p>

```python
df.columns
```

<h1>State</h1>

```python
#plotting a bar chart between the orders and the state 
state_orders= df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by="Orders",ascending=False).head(10)
sns.set(rc={"figure.figsize":(15,5)})
sns.barplot(x="State",y="Orders",data=state_orders
            ,hue="State"
           )
```

```python
#plotting a bar chart between states and the amount 
states_amount=df.groupby(["State"],as_index=False)['Amount'].sum().sort_values(by="Amount",ascending=False).head(10) 
sns.set(rc={"figure.figsize":(15,5)})
sns.barplot(x="State",y="Amount",data = states_amount ,hue="State") 
```

<p>From the above graph we can see that the most of the total sales/amount is from Uttar Pradesh, Maharashtra and Karnataka states.</p>

```python
df.columns
```

<h1>Marital Status</h1>

```python
ax=sns.countplot(x="Marital_Status",data=df,hue="Marital_Status") 
sns.set(rc={'figure.figsize':(7,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#plotting a bar chart between marital status , gender vs amount
maritalstatus_amount=df.groupby(['Marital_Status','Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False) 
sns.set({"figure.figsize":(6,5)})
sns.barplot(x="Marital_Status",y="Amount",hue="Gender",data=maritalstatus_amount)
```

<p>From the above graphs we can see that the most of the amount/total salers are from married women</p>

<h1>Occupation</h1>

```python
sns.set(rc=({'figure.figsize':(20,5)}))
ax=sns.countplot(x="Occupation",data=df,hue="Occupation") 
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#plotting a bar chart between occupation and amount
occupation_amount=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False) 
sns.set(rc={'figure.figsize':(20,5)}) 
sns.barplot(x="Occupation",y="Amount",hue="Occupation", data= occupation_amount)
```

<p>From the above graph we can see that most of the total sales / amount is from the IT sector , Health Care and Aviation Occupations</p>

<h1>Product Category</h1>

```python
sns.set(rc={'figure.figsize':(25,10)})
ax=sns.countplot(data=df,x="Product_Category",hue="Product_Category") 
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#plotting a bar chart between product categories and amount 
productcategories_amount=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by="Amount",ascending=False).head(10)
sns.set(rc=({'figure.figsize':(20,10)})) 
sns.barplot(x='Product_Category',y='Amount',data=productcategories_amount , hue="Product_Category")
```

<p>From the above graph we can see that the most sales are from food,clothing & apparel, Electronics and gadget categories</p>

```python
#top 10 selling products 
fig1,ax1 =plt.subplots(figsize=(12,7)) 
df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(kind="bar")
```
